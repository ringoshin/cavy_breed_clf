```python
# fix random seed for reproducibility
seed = 128
np.random.seed(seed)

# Image size for standardisation
image_shape = (150, 150)
input_shape = (-1, 150, 150, 3)
```

```python
new_samples_path = 'images/new_samples/'
new_samples_list = os.listdir('images/new_samples')

nrows = 150
ncolumns = 150
channels = 3
batch_size =64
```

```python
def Read_and_Process_My_Images(list_of_images, X, y):
    global nrows, ncolumns, new_samples_path
    
    for image in list_of_images:
        image=new_samples_path+image
            
        #print(image)
        X.append(cv2.resize(cv2.imread(image, cv2.IMREAD_COLOR), \
                            (nrows, ncolumns), interpolation=cv2.INTER_CUBIC))
        y.append(image)
    return X, y
```

```python
def Read_and_Process_One_Image(image_file):
    global nrows, ncolumns, new_samples_path
    
    image=new_samples_path+image_file
            
    #print(image)
    X = (cv2.resize(cv2.imread(image, cv2.IMREAD_COLOR), \
                            (nrows, ncolumns), interpolation=cv2.INTER_CUBIC))
    y = image
    return X, y
```
